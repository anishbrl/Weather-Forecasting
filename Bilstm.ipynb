```python
seq_length=24
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_train=[]
Y_train=[]

for i in range(seq_length,training_data.shape[0]):
    X_train.append(training_data[i-seq_length:i])
    Y_train.append(training_data[i,0])
X_train,Y_train=np.array(X_train),np.array(Y_train)
Y_train=Y_train.reshape(-1,1)
print(X_train.shape,Y_train.shape)
X_train = Variable(torch.Tensor(X_train))
Y_train = Variable(torch.Tensor(Y_train))
print(X_train.shape,Y_train.shape)
```

```python
class BLSTM(nn.Module):
    def __init__(self, num_classes, input_size, hidden_size, num_layers):
        super(BLSTM, self).__init__()

        self.num_classes = num_classes
        self.num_layers = num_layers
        self.input_size = input_size
        self.hidden_size = hidden_size

        self.blstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size,
                             num_layers=num_layers, batch_first=True, bidirectional=True).to(device)

        self.fc = nn.Linear(hidden_size * 2, num_classes)

    def forward(self, x):
        h_0 = torch.randn(self.num_layers * 2, x.size(0), self.hidden_size).requires_grad_().to(device)
        c_0 = torch.randn(self.num_layers * 2, x.size(0), self.hidden_size).requires_grad_().to(device)

        output, (h_n, c_n) = self.blstm(x, (h_0, c_0))

        out = self.fc(output[:, -1, :])
        return out
```

```python
num_epochs = 1000
learning_rate = 0.01
input_size = 1
hidden_size = 4
hidden_size2= 4
num_layers = 1
num_classes = 1
lstm = BLSTM(num_classes, input_size, hidden_size,num_layers)
lstm.to(device)

print(lstm)
```

```python
# DEFINE OPTIMIZER
criterion = torch.nn.MSELoss()    # mean-squared error for regression
optimizer = torch.optim.Adam(lstm.parameters(), lr=learning_rate)
```

```python
past_60_days = data_training.tail(60)
df = pd.concat([past_60_days, data_test], ignore_index=True)
df = data_training[['PRECTOT']]
print(df.shape)
```

```python
import matplotlib.pyplot as plt

losses = []

for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = lstm(X_train.to(device))

    # obtain the loss function
    loss = criterion(outputs, Y_train.to(device))
    losses.append(loss.item())

    loss.backward()
    optimizer.step()

    if epoch % 50 == 0:
        print("Epoch: %d, loss: %1.5f" % (epoch, loss.item()))

# Plot epoch vs loss curve
plt.plot(range(num_epochs), losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Epoch vs Loss Curve')
plt.show()
```

```python
import torch
import torch.nn as nn

# Define the BLSTM model class
class BLSTM(nn.Module):
    def __init__(self):
        super(BLSTM, self).__init__()
        self.blstm = nn.LSTM(1, 4, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(8, 1)

    def forward(self, x):
        lstm_out, _ = self.blstm(x)
        output = self.fc(lstm_out[:, -1, :])  # Use only the last time step for forecasting
        return output

# Define the forecasting function
def forecast_temperature(trained_model, present_temperature):
    # Load the trained model
    model = BLSTM()
    model.load_state_dict(torch.load(trained_model, map_location=torch.device('cpu')))
    model.eval()

    # Prepare the input tensor for forecasting
    with torch.no_grad():
        input_tensor = torch.tensor(present_temperature).unsqueeze(0).unsqueeze(-1).float()

        # Perform the forecast
        future_temperature = []
        for _ in range(6):  # Forecasting future 6 days
            forecasted_temp = model(input_tensor)
            future_temperature.append(forecasted_temp.item())
            input_tensor = torch.cat([input_tensor[:, 1:, :], forecasted_temp.unsqueeze(-1)], dim=1)

    return future_temperature

# Example usage:
# Assuming the trained model is saved in "trained_blstm_model.pt"
present_temperature = [5.2, 4.8, 3.5, 2.9, 6.1, 5.3, 3.7, 4.9, 4.0, 2.5, 3.1, 5.8, 4.2, 2.7, 5.7, 5.1, 6.2, 3.8, 4.3, 5.0, 3.3, 6.0, 2.6, 4.5]

scaling_factor = 0.01221598

# Inverse scale the forecasted temperature values
present_temperature = [value * scaling_factor for value in present_temperature]
forecasted_temperature = forecast_temperature("model.pt", present_temperature)
print("Forecasted temperature for the next 6 days:")
print(forecasted_temperature)
```
